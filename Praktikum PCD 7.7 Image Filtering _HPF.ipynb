```python
# Highpass Filter

# sebenarnya kita tidak perlu melakukan filtering lagi. Cukup sekali saja 
# di bagian awal, selama notebook ini tetap terhubung
import cv2
import numpy as np
from matplotlib import pyplot as plt


# memanggil citra sebagai grayscale (argument 0)
img = cv2.imread('ludo.jpg',0)

# menerapkan algoritma high-pass filtering:
# laplacian
laplacian = cv2.Laplacian(img,cv2.CV_64F)

# sobel dengan ukuran kernel 5
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

# Catatan:
# CV_64F pada contoh di atas menunjukkan nilai bit dari citra 
# yang dihasilkan serta tipe datanya (F = Float)

# perbesar ukuran hasil plotting 
plt.rcParams["figure.figsize"] = (20,20)


# menampilkan hasil filter
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Memanggil citra sebagai grayscale (argument 0)
img = cv2.imread('ludo.jpg', 0)

# Menerapkan algoritma high-pass filtering: Laplacian
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Menghitung histogram dari citra asli
hist_img, bins_img = np.histogram(img.flatten(), bins=256, range=[0, 256])

# Menghitung histogram dari citra hasil filter: Laplacian
hist_laplacian, bins_laplacian = np.histogram(laplacian.flatten(), bins=256, range=[0, 256])

# Menerapkan algoritma high-pass filtering: Sobel X
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)

# Menghitung histogram dari citra hasil filter: Sobel X
hist_sobelx, bins_sobelx = np.histogram(sobelx.flatten(), bins=256, range=[0, 256])

# Menerapkan algoritma high-pass filtering: Sobel Y
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

# Menghitung histogram dari citra hasil filter: Sobel Y
hist_sobely, bins_sobely = np.histogram(sobely.flatten(), bins=256, range=[0, 256])

# Membuat subplot grid dengan tampilan 2 baris dan 4 kolom
fig, axs = plt.subplots(2, 4, figsize=(16, 8))

# Menampilkan citra dan histogram pada setiap subplot
axs[0, 0].imshow(img, cmap='gray')
axs[0, 0].set_title('Original')
axs[0, 0].axis('off')
axs[0, 1].imshow(laplacian, cmap='gray')
axs[0, 1].set_title('Laplacian')
axs[0, 1].axis('off')
axs[0, 2].imshow(sobelx, cmap='gray')
axs[0, 2].set_title('Sobel X')
axs[0, 2].axis('off')
axs[0, 3].imshow(sobely, cmap='gray')
axs[0, 3].set_title('Sobel Y')
axs[0, 3].axis('off')

axs[1, 0].plot(hist_img)
axs[1, 0].set_title('Original Histogram')
axs[1, 0].set_xlabel('Pixel Value')
axs[1, 0].set_ylabel('Frequency')
axs[1, 1].plot(hist_laplacian)
axs[1, 1].set_title('Laplacian Histogram')
axs[1, 1].set_xlabel('Pixel Value')
axs[1, 1].set_ylabel('Frequency')
axs[1, 2].plot(hist_sobelx)
axs[1, 2].set_title('Sobel X Histogram')
axs[1, 2].set_xlabel('Pixel Value')
axs[1, 2].set_ylabel('Frequency')
axs[1, 3].plot(hist_sobely)
axs[1, 3].set_title('Sobel Y Histogram')
axs[1, 3].set_xlabel('Pixel Value')
axs[1, 3].set_ylabel('Frequency')

# Menampilkan subplot grid
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Memanggil citra sebagai grayscale (argument 0)
img = cv2.imread('turningRed.jpg', 0)

# Memanggil fungsi Canny Edges dengan argument (citra, nilai_min, nilai_max)
edges = cv2.Canny(img, 100, 200)

# Menghitung histogram dari citra asli
hist_img, bins_img = np.histogram(img.flatten(), bins=256, range=[0, 256])

# Menghitung histogram dari citra hasil Canny Edges
hist_edges, bins_edges = np.histogram(edges.flatten(), bins=256, range=[0, 256])

# Membuat subplot grid dengan tampilan 2 baris dan 2 kolom
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Menampilkan citra asli pada subplot pertama
axs[0, 0].imshow(img, cmap='gray')
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

# Menampilkan citra hasil Canny Edges pada subplot kedua
axs[0, 1].imshow(edges, cmap='gray')
axs[0, 1].set_title('Edge Image')
axs[0, 1].axis('off')

# Menampilkan histogram citra asli pada subplot ketiga
axs[1, 0].bar(range(256), hist_img, width=1.0, color='b')
axs[1, 0].set_title('Original Image Histogram')
axs[1, 0].set_xlabel('Pixel Value')
axs[1, 0].set_ylabel('Frequency')

# Menampilkan histogram citra hasil Canny Edges pada subplot keempat
axs[1, 1].bar(range(256), hist_edges, width=1.0, color='b')
axs[1, 1].set_title('Edge Image Histogram')
axs[1, 1].set_xlabel('Pixel Value')
axs[1, 1].set_ylabel('Frequency')

# Menampilkan subplot grid
plt.tight_layout()
plt.show()
```
