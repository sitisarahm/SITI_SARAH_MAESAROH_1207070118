```python
# Mengimpor modul matplotlib.pyplot dengan alias plt untuk digunakan dalam visualisasi data
import matplotlib.pyplot as plt

# Menggunakan perintah "%matplotlib inline" untuk menampilkan plot di dalam notebook secara langsung
%matplotlib inline

# Mengimpor modul skimage.data untuk mengakses koleksi gambar contoh yang disediakan oleh scikit-image
from skimage import data

# Mengimpor fungsi imread dari modul skimage.io untuk membaca gambar dari file
from skimage.io import imread

# Mengimpor fungsi rgb2gray dari modul skimage.color untuk mengubah gambar berwarna menjadi skala keabuan
from skimage.color import rgb2gray 

# Mengimpor modul numpy dengan alias np untuk digunakan dalam manipulasi array numerik
import numpy as np
```

```python
# Membaca citra "mobil.tif" menggunakan fungsi imread dari modul skimage.io dan menyimpannya dalam variabel citra1
citra1 = imread(fname="mobil.tif")

# Membaca citra "boneka2.tif" menggunakan fungsi imread dari modul skimage.io dan menyimpannya dalam variabel citra2
citra2 = imread(fname="boneka2.tif")

# Menampilkan dimensi citra 1 dengan menggunakan atribut shape dan mencetaknya
print('Shape citra 1 : ', citra1.shape)

# Menampilkan dimensi citra 2 dengan menggunakan atribut shape dan mencetaknya
print('Shape citra 2 : ', citra2.shape)

# Membuat plot berukuran 1x2 dengan ukuran gambar 10x10 menggunakan fungsi subplots dari modul matplotlib.pyplot
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

# Membuat array dari objek sumbu menggunakan fungsi ravel dari modul numpy
ax = axes.ravel()

# Menampilkan citra 1 pada sumbu pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra 1")

# Menampilkan citra 2 pada sumbu kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Citra 2")
```

```python
# Membuat salinan citra 1 dengan menggunakan metode copy() dan menyimpannya dalam variabel copyCitra1
copyCitra1 = citra1.copy()

# Membuat salinan citra 2 dengan menggunakan metode copy() dan menyimpannya dalam variabel copyCitra2
copyCitra2 = citra2.copy()

# Mendapatkan dimensi citra 1 dengan menggunakan atribut shape dan menyimpan ukuran baris dan kolom dalam variabel m1 dan n1
m1, n1 = copyCitra1.shape

# Membuat array kosong dengan ukuran m1 x n1 menggunakan fungsi empty dari modul numpy dan menyimpannya dalam variabel output1
output1 = np.empty([m1, n1])

# Mendapatkan dimensi citra 2 dengan menggunakan atribut shape dan menyimpan ukuran baris dan kolom dalam variabel m2 dan n2
m2, n2 = copyCitra2.shape

# Membuat array kosong dengan ukuran m2 x n2 menggunakan fungsi empty dari modul numpy dan menyimpannya dalam variabel output2
output2 = np.empty([m2, n2])

# Menampilkan dimensi salinan citra 1 dengan menggunakan atribut shape dan mencetaknya
print('Shape copy citra 1 : ', copyCitra1.shape)

# Menampilkan dimensi output citra 1 dengan menggunakan atribut shape dan mencetaknya
print('Shape output citra 1 : ', output1.shape)

# Menampilkan nilai m1, yaitu jumlah baris citra 1, dan mencetaknya
print('m1 : ', m1)

# Menampilkan nilai n1, yaitu jumlah kolom citra 1, dan mencetaknya
print('n1 : ', n1)
print()

# Menampilkan dimensi salinan citra 2 dengan menggunakan atribut shape dan mencetaknya
print('Shape copy citra 2 : ', copyCitra2.shape)

# Menampilkan dimensi output citra 2 dengan menggunakan atribut shape dan mencetaknya
print('Shape output citra 2 : ', output2.shape)

# Menampilkan nilai m2, yaitu jumlah baris citra 2, dan mencetaknya
print('m2 : ', m2)

# Menampilkan nilai n2, yaitu jumlah kolom citra 2, dan mencetaknya
print('n2 : ', n2)
print()
```

```python
# Melakukan iterasi untuk setiap baris dalam range 0 hingga m1-1 pada citra 1
for baris in range(0, m1-1):
    # Melakukan iterasi untuk setiap kolom dalam range 0 hingga n1-1 pada citra 1
    for kolom in range(0, n1-1):
        
        # Menyimpan indeks baris dalam variabel a1
        a1 = baris
        
        # Menyimpan indeks kolom dalam variabel b1
        b1 = kolom
        
        # Membuat array arr yang berisi piksel sekitar dari citra 1 dengan menggunakan indeks a1 dan b1
        arr = np.array([copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1, b1+1], \
            copyCitra1[a1, b1-1], copyCitra1[a1, b1+1], copyCitra1[a1+1, b1-1],  \
            copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]])
        
        # Menentukan piksel terkecil dalam array arr menggunakan fungsi amin dari modul numpy dan menyimpannya dalam variabel minPiksel
        minPiksel = np.amin(arr)
        
        # Menentukan piksel terbesar dalam array arr menggunakan fungsi amax dari modul numpy dan menyimpannya dalam variabel maksPiksel
        maksPiksel = np.amax(arr)
        
        # Memeriksa apakah piksel pada posisi baris, kolom pada citra 1 lebih kecil dari piksel terkecil
        if copyCitra1[baris, kolom] < minPiksel:
            # Jika iya, maka piksel pada posisi tersebut diubah menjadi piksel terkecil
            output1[baris, kolom] = minPiksel
        else:
            # Jika tidak, maka memeriksa apakah piksel pada posisi baris, kolom pada citra 1 lebih besar dari piksel terbesar
            if copyCitra1[baris, kolom] > maksPiksel:
                # Jika iya, maka piksel pada posisi tersebut diubah menjadi piksel terbesar
                output1[baris, kolom] = maksPiksel
            else:
                # Jika tidak, maka piksel pada posisi tersebut tetap sama
                output1[baris, kolom] = copyCitra1[baris, kolom]
```

```python
# Melakukan iterasi untuk setiap baris dalam range 0 hingga m2-1 pada citra 2
for baris1 in range(0, m2-1):
    # Melakukan iterasi untuk setiap kolom dalam range 0 hingga n2-1 pada citra 2
    for kolom1 in range(0, n2-1):
        
        # Menyimpan indeks baris dalam variabel a1
        a1 = baris1
        
        # Menyimpan indeks kolom dalam variabel b1
        b1 = kolom1
        
        # Membuat array arr yang berisi piksel sekitar dari citra 2 dengan menggunakan indeks a1 dan b1
        arr = np.array([copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1, b1+1], \
            copyCitra2[a1, b1-1], copyCitra2[a1, b1+1], copyCitra2[a1+1, b1-1],  \
            copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]])
        
        # Menentukan piksel terkecil dalam array arr menggunakan fungsi amin dari modul numpy dan menyimpannya dalam variabel minPiksel
        minPiksel = np.amin(arr)
        
        # Menentukan piksel terbesar dalam array arr menggunakan fungsi amax dari modul numpy dan menyimpannya dalam variabel maksPiksel
        maksPiksel = np.amax(arr)
        
        # Memeriksa apakah piksel pada posisi baris1, kolom1 pada citra 2 lebih kecil dari piksel terkecil
        if copyCitra2[baris1, kolom1] < minPiksel:
            # Jika iya, maka piksel pada posisi tersebut diubah menjadi piksel terkecil
            output2[baris1, kolom1] = minPiksel
        else:
            # Jika tidak, maka memeriksa apakah piksel pada posisi baris1, kolom1 pada citra 2 lebih besar dari piksel terbesar
            if copyCitra2[baris1, kolom1] > maksPiksel:
                # Jika iya, maka piksel pada posisi tersebut diubah menjadi piksel terbesar
                output2[baris1, kolom1] = maksPiksel
            else:
                # Jika tidak, maka piksel pada posisi tersebut tetap sama
                output2[baris1, kolom1] = copyCitra2[baris1, kolom1]
```

```python
# Membuat sebuah figure dengan 2 baris dan 2 kolom menggunakan plt.subplots
# dengan ukuran figur sebesar 10x10
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Menggabungkan semua sumbu menjadi satu dimensi menggunakan ravel()
ax = axes.ravel()

# Menampilkan citra 1 pada sumbu pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Input Citra 1")

# Menampilkan citra 2 pada sumbu kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Input Citra 2")

# Menampilkan output citra 1 pada sumbu ketiga
ax[2].imshow(output1, cmap='gray')
ax[2].set_title("Output Citra 1")

# Menampilkan output citra 2 pada sumbu keempat
ax[3].imshow(output2, cmap='gray')
ax[3].set_title("Output Citra 2")
```
