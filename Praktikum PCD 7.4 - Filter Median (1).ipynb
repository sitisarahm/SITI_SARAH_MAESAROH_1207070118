```python
# Mengimpor modul matplotlib.pyplot dengan alias plt
import matplotlib.pyplot as plt

# Magic command untuk menampilkan plot di dalam notebook yang sama
%matplotlib inline

# Mengimpor modul data dari paket skimage
from skimage import data

# Mengimpor fungsi imread dari modul io dalam paket skimage
from skimage.io import imread

# Mengimpor fungsi rgb2gray dari modul color dalam paket skimage
from skimage.color import rgb2gray 

# Mengimpor modul numpy dengan alias np
import numpy as np
```

```python
# Membaca citra mobil.tif dan menyimpannya ke dalam variabel citra1
citra1 = imread(fname="mobil.tif")

# Membaca citra boneka2.tif dan menyimpannya ke dalam variabel citra2
citra2 = imread(fname="boneka2.tif")

# Menampilkan dimensi citra1
print('Shape citra 1 : ', citra1.shape)

# Menampilkan dimensi citra2
print('Shape citra 2 : ', citra2.shape)

# Membuat subplot dengan 1 baris dan 2 kolom, dengan ukuran figure 10x10
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

# Membuat array ax yang berisi subplot
ax = axes.ravel()

# Menampilkan citra1 di subplot pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra 1")

# Menampilkan citra2 di subplot kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Citra 2")
```

```python
# Meng-copy citra1 ke variabel copyCitra1
copyCitra1 = citra1.copy()

# Meng-copy citra2 ke variabel copyCitra2
copyCitra2 = citra2.copy()

# Mendapatkan dimensi citra1
m1, n1 = copyCitra1.shape

# Membuat array kosong dengan ukuran m1 x n1
output1 = np.empty([m1, n1])

# Mendapatkan dimensi citra2
m2, n2 = copyCitra2.shape

# Membuat array kosong dengan ukuran m2 x n2
output2 = np.empty([m2, n2])

# Menampilkan dimensi copyCitra1
print('Shape copy citra 1 : ', copyCitra1.shape)

# Menampilkan dimensi output1
print('Shape output citra 1 : ', output1.shape)

# Menampilkan nilai m1
print('m1 : ', m1)

# Menampilkan nilai n1
print('n1 : ', n1)
print()

# Menampilkan dimensi copyCitra2
print('Shape copy citra 2 : ', copyCitra2.shape)

# Menampilkan dimensi output2
print('Shape output citra 2 : ', output2.shape)

# Menampilkan nilai m2
print('m2 : ', m2)

# Menampilkan nilai n2
print('n2 : ', n2)
print()
```

```python
# Loop untuk setiap baris dalam range 0 hingga m1-1
for baris in range(0, m1-1):
    # Loop untuk setiap kolom dalam range 0 hingga n1-1
    for kolom in range(0, n1-1):
        # Mendefinisikan variabel a1 dan b1 dengan nilai baris dan kolom saat ini
        a1 = baris
        b1 = kolom
        
        # Mendapatkan data sekitar citra pada posisi tertentu
        dataA = [copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1-1, b1+1], \
                  copyCitra1[a1, b1-1], copyCitra1[a1, b1], copyCitra1[a1, b1+1], \
                  copyCitra1[a1+1, b1-1], copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]]
        
        # Mengurutkan dataA dalam urutan naik
        for i in range(1, 8):
            for j in range(i, 9):
                if dataA[i] > dataA[j]:
                    tmpA = dataA[i]
                    dataA[i] = dataA[j]
                    dataA[j] = tmpA
        
        # Menyimpan nilai dataA[5] (nilai tengah setelah diurutkan) ke dalam output1[a1, b1]
        output1[a1, b1] = dataA[5]
```

```python
# Loop untuk setiap baris dalam range 0 hingga m2-1
for baris in range(0, m2-1):
    # Loop untuk setiap kolom dalam range 0 hingga n2-1
    for kolom in range(0, n2-1):
        # Mendefinisikan variabel a1 dan b1 dengan nilai baris dan kolom saat ini
        a1 = baris
        b1 = kolom
        
        # Mendapatkan data sekitar citra pada posisi tertentu
        dataA = [copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1-1, b1+1], \
                  copyCitra2[a1, b1-1], copyCitra2[a1, b1], copyCitra2[a1, b1+1], \
                  copyCitra2[a1+1, b1-1], copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]]
        
        # Mengurutkan dataA dalam urutan naik
        for i in range(1, 8):
            for j in range(i, 9):
                if dataA[i] > dataA[j]:
                    tmpA = dataA[i]
                    dataA[i] = dataA[j]
                    dataA[j] = tmpA
        
        # Menyimpan nilai dataA[5] (nilai tengah setelah diurutkan) ke dalam output2[a1, b1]
        output2[a1, b1] = dataA[5]
```

```python
# Membuat subplot dengan 2 baris dan 2 kolom, dengan ukuran figure 10x10
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Membuat array ax yang berisi subplot
ax = axes.ravel()

# Menampilkan citra1 di subplot pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Input Citra 1")

# Menampilkan citra2 di subplot kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Input Citra 2")

# Menampilkan output1 di subplot ketiga
ax[2].imshow(output1, cmap='gray')
ax[2].set_title("Output Citra 1")

# Menampilkan output2 di subplot keempat
ax[3].imshow(output2, cmap='gray')
ax[3].set_title("Output Citra 2")
```
