```python
# Mengimpor modul matplotlib.pyplot dengan alias plt. Modul ini digunakan untuk membuat visualisasi data dalam bentuk grafik.
import matplotlib.pyplot as plt

# Mengaktifkan visualisasi grafik agar ditampilkan secara langsung di dalam notebook.
%matplotlib inline

# Mengimpor modul data dari pustaka skimage. Skimage adalah pustaka Python yang digunakan untuk memproses gambar.
from skimage import data

# Mengimpor fungsi imread dari modul io dalam pustaka skimage. Fungsi ini digunakan untuk membaca gambar dari file.
from skimage.io import imread

# Mengimpor fungsi rgb2gray dari modul color dalam pustaka skimage. Fungsi ini digunakan untuk mengkonversi gambar berwarna menjadi gambar grayscale.
from skimage.color import rgb2gray

# Mengimpor modul numpy dengan alias np. Numpy adalah pustaka Python yang digunakan untuk melakukan operasi numerik dan manipulasi array.
import numpy as np
```

```python
# Membaca gambar dengan nama file "mobil.tif" dan menyimpannya ke variabel `citra1`.
citra1 = imread(fname="mobil.tif")

# Membaca gambar dengan nama file "boneka2.tif" dan menyimpannya ke variabel `citra2`.
citra2 = imread(fname="boneka2.tif")

# Menampilkan dimensi (shape) dari citra1 menggunakan atribut `.shape`.
print('Shape citra 1 : ', citra1.shape)

# Menampilkan dimensi (shape) dari citra2 menggunakan atribut `.shape`.
print('Shape citra 2 : ', citra2.shape)

# Membuat sebuah figure dan dua sumbu (axes) menggunakan `subplots()`. Figure ini akan menampilkan dua gambar secara berdampingan.
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

# Mengubah array sumbu `axes` menjadi array 1 dimensi menggunakan `.ravel()`.
ax = axes.ravel()

# Menampilkan citra1 pada sumbu pertama (`ax[0]`) menggunakan `.imshow()`. Citra ditampilkan dalam skala abu-abu (grayscale) dengan colormap 'gray'.
ax[0].imshow(citra1, cmap='gray')

# Memberikan judul "Citra 1" pada sumbu pertama (`ax[0]`) menggunakan `.set_title()`.
ax[0].set_title("Citra 1")

# Menampilkan citra2 pada sumbu kedua (`ax[1]`) menggunakan `.imshow()`. Citra ditampilkan dalam skala abu-abu (grayscale) dengan colormap 'gray'.
ax[1].imshow(citra2, cmap='gray')

# Memberikan judul "Citra 2" pada sumbu kedua (`ax[1]`) menggunakan `.set_title()`.
ax[1].set_title("Citra 2")
```

```python
# Menggunakan metode `.copy()` untuk membuat salinan dari citra1 dan mengkonversinya ke tipe data float menggunakan `.astype(float)`.
copyCitra1 = citra1.copy().astype(float)

# Menggunakan metode `.copy()` untuk membuat salinan dari citra2 dan mengkonversinya ke tipe data float menggunakan `.astype(float)`.
copyCitra2 = citra2.copy().astype(float)

# Menyimpan dimensi citra1 (jumlah baris dan kolom) ke variabel `m1` dan `n1`.
m1, n1 = copyCitra1.shape

# Membuat array kosong dengan dimensi yang sama dengan citra1 menggunakan `np.empty()` dan menyimpannya ke variabel `output1`.
output1 = np.empty([m1, n1])

# Menyimpan dimensi citra2 (jumlah baris dan kolom) ke variabel `m2` dan `n2`.
m2, n2 = copyCitra2.shape

# Membuat array kosong dengan dimensi yang sama dengan citra2 menggunakan `np.empty()` dan menyimpannya ke variabel `output2`.
output2 = np.empty([m2, n2])

# Menampilkan dimensi (shape) dari copyCitra1 menggunakan atribut `.shape`.
print('Shape copy citra 1 : ', copyCitra1.shape)

# Menampilkan dimensi (shape) dari output1 menggunakan atribut `.shape`.
print('Shape output citra 1 : ', output1.shape)

# Menampilkan nilai dari variabel `m1`.
print('m1 : ', m1)

# Menampilkan nilai dari variabel `n1`.
print('n1 : ', n1)

print()

# Menampilkan dimensi (shape) dari copyCitra2 menggunakan atribut `.shape`.
print('Shape copy citra 2 : ', copyCitra2.shape)

# Menampilkan dimensi (shape) dari output2 menggunakan atribut `.shape`.
print('Shape output citra 3 : ', output2.shape)

# Menampilkan nilai dari variabel `m2`.
print('m2 : ', m2)

# Menampilkan nilai dari variabel `n2`.
print('n2 : ', n2)

print()
```

```python
# Iterasi melalui setiap baris dalam rentang 0 hingga m1-1
for baris in range(0, m1-1):
    # Iterasi melalui setiap kolom dalam rentang 0 hingga n1-1
    for kolom in range(0, n1-1):
        # Menyimpan nilai baris saat ini dalam variabel a1
        a1 = baris
        # Menyimpan nilai kolom saat ini dalam variabel b1
        b1 = kolom
        
        # Menghitung jumlah elemen sekitar lokasi (a1, b1)
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]
        
        # Menetapkan nilai rata-rata dari jumlah ke lokasi (a1, b1) dalam matriks output1
        output1[a1, b1] = (1/9 * jumlah)
```

```python
# Iterasi melalui setiap baris dalam rentang 0 hingga m2-1
for baris1 in range(0, m2-1):
    # Iterasi melalui setiap kolom dalam rentang 0 hingga n2-1
    for kolom1 in range(0, n2-1):
        # Menyimpan nilai baris saat ini dalam variabel a1
        a1 = baris1
        # Menyimpan nilai kolom saat ini dalam variabel b1
        b1 = kolom1
        
        # Menghitung jumlah elemen sekitar lokasi (a1, b1)
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1-1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1]
        
        # Menetapkan nilai rata-rata dari jumlah ke lokasi (a1, b1) dalam matriks output2
        output2[a1, b1] = (1/9 * jumlah)
```

```python
# Membuat subplot dengan ukuran 2x2 dan ukuran total gambar 10x10
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Membuat array dari subplot agar lebih mudah diakses
ax = axes.ravel()

# Menampilkan citra pertama (citra1) pada subplot indeks 0
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Input Citra 1")

# Menampilkan citra kedua (citra2) pada subplot indeks 1
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Input Citra 2")

# Menampilkan citra hasil pengolahan pertama (output1) pada subplot indeks 2
ax[2].imshow(output1, cmap='gray')
ax[2].set_title("Output Citra 1")

# Menampilkan citra hasil pengolahan kedua (output2) pada subplot indeks 3
ax[3].imshow(output2, cmap='gray')
ax[3].set_title("Output Citra 2")

# Menampilkan plot dengan subplot yang telah disiapkan
plt.show()
```
