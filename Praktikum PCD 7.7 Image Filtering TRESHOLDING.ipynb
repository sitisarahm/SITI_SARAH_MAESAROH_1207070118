# Image Tresholding

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# membaca gambar baymax
img = cv2.imread('turningRed.jpg',0)

# Hitungan threshold.
# Perhatikan nilai ambang batas bawah dan atas dari tiap fungsi
# yang diberikan
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

# menampilkan hasil
titles = ['Gambar asli','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

# mengatur ukuran gambar dan histogram
fig, axes = plt.subplots(6, 2, figsize=(12, 24))

# menampilkan gambar dan histogram satu per satu
for i in range(6):
    # menampilkan gambar
    axes[i, 0].imshow(images[i], 'gray')
    axes[i, 0].set_title(titles[i])
    axes[i, 0].axis('off')
    
    # menampilkan histogram
    axes[i, 1].hist(images[i].ravel(), 256, [0, 256])
    axes[i, 1].set_xlabel('Intensitas Piksel')
    axes[i, 1].set_ylabel('Jumlah Piksel')
    axes[i, 1].set_title('Histogram ' + titles[i])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# membaca gambar baymax
img = cv2.imread('turningRed.jpg', 0)

# digunakan median blur untuk menghaluskan tepi objek pada citra
# ini diperlukan agar thresholding memberikan hasil lebih baik
img = cv2.medianBlur(img, 5)

# Lakukan Thresholding
# Binary Threshold
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Adaptive Threshold dengan Mean
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Adaptive Threshold dengan Gaussian
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Plotting
titles = ['Original Image', 'Global Thresholding (v = 127)', 'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

# mengatur ukuran gambar dan histogram
fig, axes = plt.subplots(4, 2, figsize=(8, 12))

# menampilkan hasil dan histogram
for i in range(4):
    # menampilkan gambar
    axes[i, 0].imshow(images[i], 'gray')
    axes[i, 0].set_title(titles[i])
    axes[i, 0].axis('off')
    
    # menampilkan histogram
    axes[i, 1].hist(images[i].ravel(), 256, [0, 256])
    axes[i, 1].set_xlabel('Intensitas Piksel')
    axes[i, 1].set_ylabel('Jumlah Piksel')
    axes[i, 1].set_title('Histogram ' + titles[i])

plt.tight_layout()
plt.show()
```
