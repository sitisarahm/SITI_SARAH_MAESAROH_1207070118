```python
# Mengimpor modul matplotlib.pyplot untuk membuat visualisasi grafis
import matplotlib.pyplot as plt

# Memastikan tampilan grafis dari matplotlib ditampilkan di dalam notebook
%matplotlib inline

# Mengimpor modul data dari pustaka scikit-image (skimage) untuk mengakses koleksi gambar referensi
from skimage import data

# Mengimpor fungsi imread dari modul io di pustaka scikit-image untuk membaca gambar dari file
from skimage.io import imread

# Mengimpor fungsi rgb2gray dari modul color di pustaka scikit-image untuk mengonversi gambar RGB menjadi grayscale
from skimage.color import rgb2gray

# Mengimpor modul numpy dan memberikan alias np untuk komputasi numerik
import numpy as np

# Mengimpor modul cv2 dari pustaka OpenCV untuk pengolahan gambar dan komputer visi
import cv2
```

```python
# Membaca gambar dengan nama file "gedung.tif" menggunakan fungsi imread dan menyimpannya ke dalam variabel citra1
citra1 = imread(fname="apple.tif")

# Mencetak bentuk (shape) dari citra1 untuk mengetahui jumlah baris, jumlah kolom, dan jumlah saluran warna (jika ada)
print(citra1.shape)

# Menampilkan gambar citra1 menggunakan fungsi imshow dari modul pyplot. Argumen cmap='gray' digunakan untuk menampilkan gambar dalam skala abu-abu.
plt.imshow(citra1, cmap='gray')
```

```python
# Mendefinisikan kernel yang akan digunakan untuk operasi filter. Kernel adalah matriks (3x3) yang mempengaruhi nilai piksel dalam operasi konvolusi.
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])

# Mengaplikasikan operasi filter 2D ke citra1 menggunakan fungsi filter2D dari modul cv2. Hasilnya disimpan dalam citraOutput.
citraOutput = cv2.filter2D(citra1, -1, kernel)

fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

# Menampilkan citra input di subplot pertama. Memberikan judul "Citra Input" untuk subplot pertama.
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra Input")

# Menampilkan citra output di subplot kedua. Memberikan judul "Citra Output" untuk subplot kedua.
ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Output")
```
